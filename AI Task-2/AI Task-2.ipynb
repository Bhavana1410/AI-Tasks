<h2>Q3</h2>

```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []
z = []
add = []
for i in range(0,20):
    x.append(i)
    a=np.cos(i) 
    #f(x) = cosx
    y.append(a)
    b=np.sin(i)
    b=-b
    #f'(x) = -sinx (derivative of cosx)
    z.append(b)
    c= a+b
    #result = sum of those two functions
    add.append(c)

plt.rcParams["figure.figsize"] = [5, 5]
plt.plot(x,y)
plt.title("f(x) = cosx", fontsize = 20)
plt.xlabel("Values in radians")
plt.ylabel("Value of the function")
plt.show()
plt.plot(x,z)
plt.title("f'(x) = -sinx", fontsize = 20)
plt.xlabel("Values in radians")
plt.ylabel("Value of the function")
plt.show()
plt.plot(x,add)
plt.title("f(x)+f'(x) = cosx-sinx", fontsize = 20)
plt.xlabel("Values in radians")
plt.ylabel("Value of the function")
plt.show()
```
